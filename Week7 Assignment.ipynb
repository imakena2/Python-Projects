```python
import pandas as pd

try:
    df = pd.read_csv('aggregated_telematics_24vars.csv')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print(f"An error occurred: {e}")

# Display the first few rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna() 
print(df.columns)
```

```python
# Filter numeric columns only
numeric_df = df.select_dtypes(include='number')

# Compute the correlation matrix
correlations = numeric_df.corr()

# Show correlation with the target variables (NB_Claim and AMT_Claim)
correlations[['NB_Claim', 'AMT_Claim']].sort_values(by='NB_Claim', ascending=False)
```

```python

# Group by categorical column 'Region' and compute mean of 'NB_Claim'
group_by_region = df.groupby('Region')['NB_Claim'].mean()
print(group_by_region)
```

```python
# Group by 'Marital' and compute mean of 'AMT_Claim'
group_by_marital = df.groupby('Marital')['AMT_Claim'].mean()
print(group_by_marital)
```

```python
import matplotlib.pyplot as plt

# Example: Line chart for AMT_Claim over time (assuming 'Duration' represents time)
plt.figure(figsize=(10, 6))
plt.plot(df['Duration'], df['AMT_Claim'], color='blue', marker='o', linestyle='-', label='Claim Amount')
plt.title('Claim Amount Over Duration')
plt.xlabel('Duration')
plt.ylabel('AMT_Claim')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar chart for comparing average number of claims across regions
plt.figure(figsize=(10, 6))
group_by_region.plot(kind='bar', color='green')
plt.title('Average Number of Claims by Region')
plt.xlabel('Region')
plt.ylabel('Average NB_Claim')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for AMT_Claim distribution
plt.figure(figsize=(10, 6))
plt.hist(df['AMT_Claim'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Total Claim Amount')
plt.xlabel('AMT_Claim')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for NB_Claim vs AMT_Claim
plt.figure(figsize=(10, 6))
plt.scatter(df['NB_Claim'], df['AMT_Claim'], color='red', alpha=0.5)
plt.title('Scatter Plot of Number of Claims vs Total Claim Amount')
plt.xlabel('NB_Claim (Number of Claims)')
plt.ylabel('AMT_Claim (Total Claim Amount)')
plt.grid(True)
plt.show()
```

```python
#Regional Impact: Urban areas tend to have a slightly higher claim frequency than rural areas, but the difference is not significant.

#Marital Impact: Single individuals file more claims on average compared to married individuals.

#High Variability in Claims: Both the number and amount of claims have large standard deviations and ranges, suggesting a few individuals drive the totals. This should be taken into account when making policy decisions or drawing conclusions from the dataset.

#Outliers: There are likely outliers in both the number and amount of claims, which could be affecting the overall distribution.
```
